# Dictionary

Python dictionary is an unordered collection of items. Each item of a dictionary has a key/value pair.

Dictionaries are optimized to retrieve values when the key is known.

```python
stu1 = {'name':'Qasim',
        'email':'abc@gmail.com',
        'age':20}
print(stu1)
print(stu1['name'])
print(stu1['age'])
```

```python
stu1 = {'name':'Qasim',
        'email':'abc@gmail.com',
        'age':20,
        23:"hello",
        True:"working",
        2.5:"new",
        4:True
       }

print(stu1)

print(stu1.items())

#to iterate over keys and values in dictionary, use .items()
for key, value in stu1.items():
        print("\nKey: ", key)
        print("Value: ", value)

print("-------------------------------------")


#to iterate over keys only in dictionary, use .keys()
for key in stu1.keys():
        print("\nKey: ", key)

print("-------------------------------------")


#to iterate over values in dictionary, use .values()
for value in stu1.values():
        print("Value: ", value)




```

```python
# Slicing string (in the same way as you sliced the list)


b = "Hello, World!"
print(b[1:8])

print("------")

b = "Hello, World!"
print(b[1:8:2])

print("-----")
b = "Hello, World!"
print(b[:5])

print("--------")

b = "Hello, World!"
print(b[2:])

print("--------")

b = "Hello, World!"
print(b[-5:-2])
```

```python
# Modify string

a = "Hello, World!"
print(a.lower())

a = "Hello, World!"
print(a.upper())

# Remove whitespaces
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"

#Replace method (replace string with another string)
a = "Hello, World!"
print(a.replace("H", "J"))

# The split() method returns a list where the text between the specified separator becomes the list items.
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

```python
# Concatenate string
# manage variable a with b into c
a = "Hello"
b = "World"
c = a + b
print(c)

print("----------------")

# to add a space between a and b
a = "Hello"
b = "World"
c = a + " " + b
print(c)

```
